```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
data = np.loadtxt("state_over_time.txt")
data_F90 = np.loadtxt("state_over_time_F90.txt")
f = open("state_over_time.txt", "r")
line = f.readline()
elements=[]


for word in line.split(): 
    elements.append(word)

f.close()
```

```python
type(data)
```

```python
data.shape
data[:, 1]
data_F90.shape
cmap = plt.get_cmap('prism')
N=20
```

```python
plt.figure(figsize=(20,10))
for x in range(2, (len(data[0]))):
    #for y in range(2, len(elements)):
    color = cmap(float(x)/N)
    plt.plot(data[:, 0], data[:, x], c=color, label = elements[x+1])
    plt.plot(data[:, 0], data_F90[:, x], '--', c=color)
    plt.yscale('log')
    plt.ylabel("Mass Fraction")
    plt.xlabel("Time")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
```

```python
for x in range(2, (len(data[0]))):
    plt.plot(data[:, 0], data_F90[:, x], label = elements[x+1])
    plt.yscale('log')
plt.legend(bbox_to_anchor=(1.05,1), loc='upper left', borderaxespad=0)
```
